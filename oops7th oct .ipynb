```python
class Bankaccount():
    def __init__(self,account_no,acc_holdername,account_balance):
        self.account_no = account_no
        self.acc_holdername = acc_holdername
        self.account_balance= account_balance
        
    def deposit(self,amount):
        if amount>0:
            self.account_balance = self.account_balance + amount
            return f"{amount} deposited in the bank account and the new balance is {self.account_balance}"
        else:
            return "invalid entry"
        
    def withdrow(self,amount):
        if amount > 0 and amount <= self.account_balance:
            self.account_balance = self.account_balance-amount
            return f"{amount} withdrown, new balance is {self.account_balance}"
        else:
            return "invalid entry"
            
    def details(self):
        return f"account balance:{self.account_balance},account holder name: {self.acc_holdername},account no: {self.account_no}"

if __name__ =="__main__":
    account = Bankaccount('839459','shubham',5500)
    print(account.details())
    
    print(account.deposit(300))
    print(account.withdrow(250))
    
    print("current balance",account.details())
```

```python
class VehicleRentalSystem:
    def __init__(self):
        self.available_vehicles = {}
        self.rented_vehicles = {}

    def add_vehicle(self, vehicle_id, vehicle_type):
        if vehicle_id not in self.available_vehicles:
            self.available_vehicles[vehicle_id] = vehicle_type
            print(f"Added {vehicle_type} with ID {vehicle_id} to the available vehicles.")
        else:
            print(f"Vehicle with ID {vehicle_id} already exists in the available vehicles.")

    def rent_vehicle(self, vehicle_id, customer_name):
        if vehicle_id in self.available_vehicles:
            vehicle_type = self.available_vehicles.pop(vehicle_id)
            self.rented_vehicles[vehicle_id] = customer_name
            print(f"{customer_name} rented a {vehicle_type} with ID {vehicle_id}.")
        else:
            print(f"Vehicle with ID {vehicle_id} is not available for rent.")

    def return_vehicle(self, vehicle_id):
        if vehicle_id in self.rented_vehicles:
            customer_name = self.rented_vehicles.pop(vehicle_id)
            vehicle_type = self.available_vehicles[vehicle_id]  # Get the vehicle type from the rented vehicles dictionary
            self.available_vehicles[vehicle_id] = vehicle_type
            print(f"{customer_name} returned a {vehicle_type} with ID {vehicle_id}.")
        else:
            print(f"Vehicle with ID {vehicle_id} is not currently rented.")

    def display_available_vehicles(self):
        print("Available Vehicles:")
        for vehicle_id, vehicle_type in self.available_vehicles.items():
            print(f"ID: {vehicle_id}, Type: {vehicle_type}")

if __name__ == "__main__":
    rental_system = VehicleRentalSystem()

    rental_system.add_vehicle("001", "Sedan")
    rental_system.add_vehicle("002", "SUV")

    rental_system.rent_vehicle("002", "John")
    rental_system.rent_vehicle("003", "Alice")

    rental_system.display_available_vehicles()

    rental_system.return_vehicle("001")

    rental_system.display_available_vehicles()
```

```python
"""Problem 5: Product Inventory Create classes representing a product and an inventory system. 
Implement methods to add products to the inventory, update product quantity, and display available products."""


class Product:
    def __init__(self,product_name, quantity,price):
        self.product_name = product_name
        self.quantity = quantity
        self.price = price
        self.is_available = True
    
    
    def __str__(self):
        return f"the product  {self.product_name} has quantity {self.quantity} and its price is {self.price}"
    
    
class inventory:
    
    def __init__(self):
        self.newproducts  = []
        
    def addpro(self,products):
        self.newproducts.append(products)
        
    def upd(self,product_name,new_quantity):
        for products in self.newproducts:
            if products.product_name == product_name:
                products.quantity = new_quantity
                return True
        return False 
    
    def datails(self):
        avail = [products for products in self.newproducts  if products.quantity > 0 ]
        if avail:
            print("the product is available")
            for products in avail:
                print(products)
        else:
            print("the product is not available")
            
            
if __name__  == "__main__":
    product1 = Product("a",5,5000)
    product2 = Product("bc",6,5840)
    product3 = Product("abc",8,6890)
    
    
    inve = inventory()
    
    inve.addpro(product1)
    inve.addpro(product2)
    inve.addpro(product3)
    
    inve.datails()
    
    
    
    update = inve.upd("a",889)
    
    if update:
        print("there is a new updation in stocks")
    else:
        print("there is no update on stock")
        
    inve.datails()
    
        
```

```python
"""Shape Calculation Create a class representing a shape with attributes like length,
width, and height. Implement methods to calculate the area and perimeter of the shape."""

class shape2d:
    def __init__(self,length,width):
        self.length = length 
        self.width = width
        
    def area(self):
        return self.length*self.width
    
    def parimeter(self):
        return self.length + self.width
    
class shape3d:
    
    def __init__(self,length,width,height):
        self.length = length 
        self.width = width
        self.height = height
        
    def volume(self):
        return self.length*self.width*self.height
    
if __name__ == "__main__":
    
    
    rectangle = shape2d(4,5)
    Area = rectangle.area()
    Parameter = rectangle.parimeter()
    print(f"AREA IS {Area} and the parameter is {Parameter}")
    
    box = shape3d(2,3,5)
    Volume = box.volume()
    print(f"the volume of the box is {Volume}")
                       
        
        
    
    


```

```python
"""Social Media Profile Create a class representing a social media profile with attributes like username and posts. 
Implement methods to add posts, display posts, and search for posts by keyword."""


class profile:
    
    def __init__(self,username):
        self.username = username
        self.posts = []
    
    def addp(self,words):
        self.posts.append(words)
        
    def display(self):
        if not self.posts:
            print(f"username: {self.username} has no posts")
        else:
            print(f"username: {self.username} has posts")
            for i, post in enumerate(self.posts,start = 1):
                print(f"{i}, {post}")
                
    def find_post(self,keyword):
        same = [post for post in self.posts if keyword in post]
        if same:
            print(f"post by {self.username} has {keyword}")
            for i, post in enumerate(same,start = 1):
                print(f"{i},{post}")
        else:
            print("not available")
        
if __name__ == "__main__":
    
    pro = profile("shubham")
    pro.addp("in a bangalore")
    pro.addp("searching for a job")
    pro.addp("not even getting the internship")
    
    
    pro.display()
    
    keyword = "bangalore"
    
    pro.find_post(keyword)
    
                
```

```python
class TODO:
    def __init__(self):
        self.tasks = {}

    def add_task(self, task, due_date):
        if due_date in self.tasks:
            self.tasks[due_date].append(task)
        else:
            self.tasks[due_date] = [task]

    def complete(self, task, due_date):
        if due_date in self.tasks and task in self.tasks[due_date]:
            self.tasks[due_date].remove(task)

    def pending(self):
        for due_date, tasks in self.tasks.items():
            if tasks:
                print(f"Due Date: {due_date}")
                for task in tasks:
                    print(f" - {task}")

if __name__ == "__main__":
    to = TODO()
    to.add_task("eating", "4:40")
    to.add_task("walking", "8:30")
    to.add_task("sleeping", "10:30")
    
    to.complete("walking", "8:30")
    
    print("Following are the pending tasks:")
    
    to.pending()

 


    
```
